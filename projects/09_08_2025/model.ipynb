```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.linear_model import RidgeClassifier
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, accuracy_score
import kaggle as kag
from kagglehub import KaggleDatasetAdapter
import kagglehub
```

```python
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "gagandeep16/car-sales",
  "Car_sales.csv",
)

print("First 5 records:", df.head())
```

```python
df.shape
```

```python
df.columns
```

```python
X = df.drop(columns=["Vehicle_type"])
y = df["Vehicle_type"]
```

```python
OHE = OneHotEncoder()
X_encoded = OHE.fit_transform(X, y)
print(X_encoded)
```

```python
LE = LabelEncoder()
y_encoded = LE.fit_transform(y)
print(y_encoded)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.25, train_size=0.75, shuffle=True)
```

```python
model = RidgeClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_absolute_error(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
print(f"MSE : {mse}\nACC : {acc}")
```
